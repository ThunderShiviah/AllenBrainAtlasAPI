```python
%matplotlib inline

import numpy as np

from skimage import io
from skimage.color import rgb2gray

from matplotlib import pyplot as plt


file1 = '../data/img/test043_TL/p1-D1-01b.jpg'
file2 = '../data/img/test043_TL/p1-F4-01b.jpg'

''' Read files and convert to grey'''
img1 = io.imread(file1 ,as_grey=True) # query image
img2 = io.imread(file2 ,as_grey=True) #training image


'''look at the loaded files'''
f, (ax0, ax1) = plt.subplots(1, 2)
ax0.imshow(img1)
ax1.imshow(img2)
```

```python
def ret(img1,img2):
    return(img1,img2)

ret(img1,img2)[0]
```

```python
from skimage.transform import pyramid_gaussian


''' Create image pyramid'''

pyramid_in = img1 # set image and downscale dimension n.
pyramid_in2 = img2 # set image and downscale dimension n.

downscale_dim = 3

pyramid = tuple(pyramid_gaussian(pyramid_in, downscale=downscale_dim))
pyramid2 = tuple(pyramid_gaussian(pyramid_in2, downscale=downscale_dim))

''' Check pyramid results '''
pyramid_layer = 3

io.imshow(pyramid[pyramid_layer])
pyramid[pyramid_layer].shape
```

```python
from skimage import img_as_float
from scipy.ndimage import gaussian_filter
from skimage.morphology import reconstruction


''' Subtract background'''

# Convert to float: Important for subtraction later which won't work with uint8
''' image 1 '''
image = img_as_float(pyramid[pyramid_layer])
image = gaussian_filter(image, 1)

seed = np.copy(image)
seed[1:-1, 1:-1] = image.min()
mask = image

dilated = reconstruction(seed, mask, method='dilation')

im_feature = image - dilated

''' image 2 '''
image2 = img_as_float(pyramid2[pyramid_layer])
image2 = gaussian_filter(image2, 1)

seed2 = np.copy(image2)
seed2[1:-1, 1:-1] = image2.min()
mask2 = image2

dilated2 = reconstruction(seed, mask, method='dilation')

im_feature2 = image2 - dilated2

'''plot the result '''

fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(8, 2.5))

ax1.imshow(image)
ax1.set_title('original image')
ax1.axis('off')

ax2.imshow(dilated, vmin=image.min(), vmax=image.max())
ax2.set_title('dilated')
ax2.axis('off')

ax3.imshow(image - dilated)
ax3.set_title('image - dilated')
ax3.axis('off')

fig.tight_layout()
```

```python
from skimage.filters import threshold_otsu, threshold_adaptive

''' Create global binary '''

bin_in = pyramid[pyramid_layer]

global_thresh = threshold_otsu(bin_in)
binary_global = bin_in > global_thresh


io.imshow(binary_global)
binary_global.shape
```

```python
from skimage import measure

contour_in = binary_global
contour_in2 = im_feature

# Find contours at a constant value of 0.8
contours = measure.find_contours(contour_in, 0.1, fully_connected='high')

contours2 = measure.find_contours(contour_in2, 0.1, fully_connected='high')

# Display the image and plot all contours found
fig, ax = plt.subplots()
ax.imshow(contour_in, interpolation='nearest', cmap=plt.cm.gray)

ax.imshow(contour_in2, interpolation='nearest', cmap=plt.cm.gray)

for n, contour in enumerate(contours):
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2)

for n, contour in enumerate(contours2):
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2)
    
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(8, 2.5))

ax.axis('image')
ax.set_xticks([])
ax.set_yticks([])

ax1.imshow(image)
ax1.set_title('original image')
ax1.axis('off')

ax2.imshow(dilated, vmin=image.min(), vmax=image.max())
ax2.set_title('dilated')
ax2.axis('off')

ax3.imshow(image - dilated)
ax3.set_title('image - dilated')
ax3.axis('off')

fig.tight_layout()
```

```python
from skimage.feature import daisy

''' Use daisy binary descriptor to extract features'''
img = img1
descs, descs_img = daisy(img, step=180, radius=58, rings=2, histograms=6,
                         orientations=8, visualize=True)

fig, ax = plt.subplots()
ax.axis('off')
ax.imshow(descs_img)
descs_num = descs.shape[0] * descs.shape[1]
ax.set_title('%i DAISY descriptors extracted:' % descs_num)
```

```python
from skimage import transform as tf
from skimage.feature import (match_descriptors, corner_harris,
                             corner_peaks, ORB, plot_matches)

'''Try orb binary descriptor using binaries created by Otsu's method.'''

descriptor_extractor = ORB(n_keypoints=10)

''' Read in images and convert to binary'''

image1 = pyramid[pyramid_layer]
thresh = threshold_otsu(image1)
binary1 = image1 > thresh

image2 = pyramid2[pyramid_layer]
thresh = threshold_otsu(image2)
binary2 = image2 > thresh


''' Extract descriptors for the original images '''


descriptor_extractor.detect_and_extract(image1)
keypoints1 = descriptor_extractor.keypoints
descriptors1 = descriptor_extractor.descriptors

descriptor_extractor.detect_and_extract(image2)
keypoints2 = descriptor_extractor.keypoints
descriptors2 = descriptor_extractor.descriptors

matches12 = match_descriptors(descriptors1, descriptors2, cross_check=True)

''' Extract descriptors from the binaries '''

descriptor_extractor.detect_and_extract(binary1)
keypoints3 = descriptor_extractor.keypoints
descriptors3 = descriptor_extractor.descriptors

descriptor_extractor.detect_and_extract(binary2)
keypoints4 = descriptor_extractor.keypoints
descriptors4 = descriptor_extractor.descriptors

matches34 = match_descriptors(descriptors3, descriptors4, cross_check=True)

''' Extract descriptors from the subtracted backgrounds'''

descriptor_extractor.detect_and_extract(im_feature)
keypoints5 = descriptor_extractor.keypoints
descriptors5 = descriptor_extractor.descriptors

descriptor_extractor.detect_and_extract(im_feature2)
keypoints6 = descriptor_extractor.keypoints
descriptors6 = descriptor_extractor.descriptors

matches56 = match_descriptors(descriptors5, descriptors6, cross_check=True)

''' Plot the images '''

fig, ax = plt.subplots(nrows=3, ncols=1)

plt.gray()

plot_matches(ax[0], image1, image2, keypoints1, keypoints2, matches12)
plot_matches(ax[1], binary1, binary2, keypoints3, keypoints4, matches34)
plot_matches(ax[2], im_feature, im_feature2, keypoints5, keypoints6, matches56)

ax[0].axis('off')
ax[1].axis('off')
ax[2].axis('off')
```
