```python
from skimage import io
from skimage import transform as tf

import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
%matplotlib inline

```

```python
# Load the two landscape photos
img0 = io.imread('../data/img/test043_TL/p1-D1-01b.jpg',as_grey=True)
img1 = io.imread('../data/img/test043_TL/p1-F4-01b.jpg',as_grey=True)

'''look at the loaded files'''
f, (ax0, ax1) = plt.subplots(1, 2)
ax0.imshow(img0)
ax1.imshow(img1)
```

```python
pyramid0 = tuple(tf.pyramid_gaussian(img0, downscale=4))
pyramid1 = tuple(tf.pyramid_gaussian(img1, downscale=4))

'''look at the loaded files'''
f, (ax0, ax1) = plt.subplots(1, 2)
ax0.imshow(pyramid0[1])
ax1.imshow(pyramid1[1])

pyramid1[1].shape
```

```python
n = 2 #downsample parameter

src = pyramid0[n]
dst = pyramid1[n]

tform = tf.estimate_transform('affine', src, dst)

image = dst
warped = tf.warp(image, inverse_map=tform.inverse, order=5) 

f, (ax0, ax1) = plt.subplots(1, 2)
ax0.imshow(src)
ax1.imshow(dst)


f, (ax2, ax3) = plt.subplots(1, 2)

ax2.imshow(src)
ax3.imshow(warped)

from skimage import measure as ms

metric = ms.structural_similarity(src, warped)

metric
```
