```python
############# Image pre-processing layer ##########################
def create_image_pyramid(stack):
    # I put in some automatic values for the definition call. I need to check that they actually work.
    """ Create image pyramid"""


    pyramid = [tuple(pyramid_gaussian(img, downscale=2)) for img in stack]

    """ Check pyramid results """
    #Also need to put some test cases that check that downscale_dim and pyramid_layer are correct values.
    layer = 3
    return([pyramid[0][layer],pyramid[1][layer]]) #there should be a better way of putting these together.
```

```python
def subtract_background_generator(stack):
    """ Subtract background"""
    for img in stack:
        # Convert to float: Important for subtraction later which won't work with uint8
        """ image 1 """
        image = gaussian_filter(img, 1)

        seed = np.copy(img)
        seed[1:-1, 1:-1] = img.min()
        mask = img

        dilated = reconstruction(seed, mask, method='dilation')

        img_background_subtracted = img - dilated

        yield(img_background_subtracted)
        
def subtract_background(stack):
    """ Subtract background"""
    subtracted = list(subtract_background_generator(stack))
    return subtracted
```

```python
subtract = subtract_background(stack)

io.imshow(subtract[0])
```

```python
def create_binary_generator(stack):
    """ Create global binary """
    for img in stack:
        global_thresh = threshold_otsu(img)
        binary_global = img > global_thresh

        yield(binary_global)
        
def create_binary(stack):
    binary = list(create_binary_generator(stack))
    return binary

    
```
