Now that we have our image, we want to get the height and width. We can do this using python's imaging library (PIL). Since our image is already loaded as pil_im we can get the width and height using the following line of code. Note: we're using the image library from PIL that we imported before. 

**Note**: we might want to get the image size without loading the image into memory. In that case refer to [this stackoverflow link](http://stackoverflow.com/questions/15800704/python-get-image-size-without-loading-image-into-memory?rq=1) for more details. 

````python
width, height = pil_im.size 
print(width, height)
````

We got 660x472px as expected. Now to to get the center x and y coordinates.

````python
from math import floor

'''takes in the width and height of an image and returns the 
center x and y coordinates as cx and cy respectively. 
'''

cx = floor(width * .5)     
cy = floor(height * .5)

print(cx, cy) #coordinates of center of the image rounded down to nearest integer.
````

We could also get the same result using list comprehension. Not sure if this is easier to read though.

````python
center = tuple([floor(.5*x) for x in (width, height)])
print(center)
````

Now we want to plug our values into a URL generator in order to make the API request to get the correct annotation. But first, below is how to request and display our relevant SVG image. Note that we don't get labels from hovering over the svg. We'll have to add that in from the metadata later.

````python
import requests
from IPython.display import SVG
'''Requests the svg annotations of an atlas image and displays the svg only.
'''

url = 'http://api.brain-map.org/api/v2/svg_download/%(AtlasImage.id)d' % {"AtlasImage.id":100960224} 

payload = {'downsample':4,'annotation':'true'}

svg_req = requests.get(url, params=payload)

## Convert the API response to an image and display.
svg_image = SVG(svg_req._content)

svg_image #prints the svg
````
