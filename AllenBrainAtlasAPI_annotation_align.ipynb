We see that scale variable in our function is given by raising 2 to downsample value as its exponent.

The scaling factor on the example image is 4, hence our scaling factor is 2^4.

````python
scale = 2**4; #16
````

Next we need to get cx and cy which are given by the following formula in the image sync code.

The formula is taking the middle of both the width and height of the image and then taking the largest integer less than or equal to the value using the floor function. Thus we see that cx and cy return the center coordinates of our image.

Let's grab our example image and the get cx and cy.

````python
import requests #for calling the api
from matplotlib.pyplot import imshow #for displaying the image
import numpy as np 
from PIL import Image 
from StringIO import StringIO

## Grab the response data from the example image api call url
r = requests.get('http://www.brain-map.org/api/v2/section_image_download/100960224?downsample=4')

## Convert the API response to an image and display.
%matplotlib inline
pil_im = Image.open(StringIO(r._content))
imshow(np.asarray(pil_im))
````

Now that we have our image, we want to get the height and width. We can do this using python's imaging library (PIL). Since our image is already loaded as pil_im we can get the width and height using the following line of code. Note: we're using the image library from PIL that we imported before. 

**Note**: we might want to get the image size without loading the image into memory. In that case refer to [this stackoverflow link](http://stackoverflow.com/questions/15800704/python-get-image-size-without-loading-image-into-memory?rq=1) for more details. 

````python
width, height = pil_im.size 
print(width, height)
````

We got 660x472px as expected. Now to to get the center x and y coordinates.

````python
from math import floor

'''takes in the width and height of an image and returns the 
center x and y coordinates as cx and cy respectively. 
'''

cx = floor(width * .5)     
cy = floor(height * .5)

print(cx, cy) #coordinates of center of the image rounded down to nearest integer.
````

We could also get the same result using list comprehension. Not sure if this is easier to read though.

````python
center = tuple([floor(.5*x) for x in (width, height)])
print(center)
````

````python
left = cx - width * 0.5 * scale
top = cy - height * 0.5 * scale;

print cx, width, scale, left, top
````

Now we want to plug our values into a URL generator in order to make the API request to get the correct annotation. But first, below is how to request and display our relevant SVG image. Note that we don't get labels from hovering over the svg. We'll have to add that in from the metadata later.

````python
import requests
from IPython.display import SVG
'''Requests the svg annotations of an atlas image and displays the svg only.
'''

url = 'http://api.brain-map.org/api/v2/svg_download/%(AtlasImage.id)d' % {"AtlasImage.id":100960224} 

payload = {'downsample':4,'annotation':'true'}

svg_req = requests.get(url, params=payload)

## Convert the API response to an image and display.
svg_image = SVG(svg_req._content)

svg_image #prints the svg
````

````python
from io import BytesIO

from IPython.core import display
from PIL import Image


def display_pil_image(im):
   """Displayhook function for PIL Images, rendered as PNG."""

   b = BytesIO()
   im.save(b, format='png')
   data = b.getvalue()

   ip_img = display.Image(data=data, format='png', embed=True)
   return ip_img._repr_png_()


# register display func with PNG formatter:
png_formatter = get_ipython().display_formatter.formatters['image/png']
dpi = png_formatter.for_type(Image.Image, display_pil_image)
````
